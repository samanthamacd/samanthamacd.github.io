```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np 
import seaborn as sns 
```

```python
iris = sns.load_dataset('iris')
diamonds = sns.load_dataset('diamonds') 
titanic = sns.load_dataset('titanic') 
```

```python
# categorical data - Iris 
species_counts = iris['species'] 
sns.barplot(x=species_counts.index, y=species_counts.values, palette = 'viridis') 
plt.title('Frequency of Iris Species')
plt.xlabel('Count')
plt.ylabel('Species')
plt.savefig('IrisBarchart.jpg')
plt.show()
```

```python
counts = iris['species'].value_counts() 
plt.figure(figsize=(8,8))
plt.pie(counts, labels=counts.index, autopct = '%1.1f%%', startangle=90, colors=sns.color_palette('Set3'))
plt.title('Pie Chart of Iris Species') 
plt.savefig('IrisPie.jpg') 
plt.show() 
```

```python
# pie chart - titanic 

survived_data = titanic[titanic['survived'] ==1] 
survived_class = survived_data['class'].value_counts() 
plt.figure(figsize=(8,8)) 
plt.pie(survived_class, labels=survived_class.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('Set2'))
plt.title('Distribution of Survivors by Class on the Titanic') 
plt.show() 
```

```python
from scipy.stats import norm

sepal_length_data = iris['sepal_length']

plt.hist(sepal_length_data, bins=20, color='skyblue', edgecolor='black', alpha=0.7, density=True)

mu, std = norm.fit(sepal_length_data)

xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, std)
plt.plot(x, p, 'k', linewidth=2)
plt.title('Histogram with Normal Distribution Fit')
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='class', y='fare', data=titanic, palette='viridis')
plt.title('Boxplot of Fare Across Passenger Classes in Titanic')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=iris, palette='Set2', s=80)
plt.title('Scatterplot of Sepal Length vs Sepal Width in Iris Dataset')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.legend()
plt.show()
```
